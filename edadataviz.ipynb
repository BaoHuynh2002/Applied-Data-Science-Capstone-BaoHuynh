```python
### TASK  8: Cast all numeric columns to `float64`
def cast_to_float64(df):
  """Casts all numeric columns in a DataFrame to float64.

  Args:
      df (pandas.DataFrame): The DataFrame to cast.

  Returns:
      pandas.DataFrame: The DataFrame with cast numeric columns.
  """
  # Select the numeric columns
  numeric_cols = df.select_dtypes(include=['int64', 'float64'])

  # Cast those columns to float64
  df[numeric_cols.columns] = numeric_cols.astype('float64')
  return df

# Apply the function to your DataFrame (assuming it's called 'df')
df = cast_to_float64(df.copy())
```
